Tipos sequências — <a class="reference internal" href="#list" title="list"><code class="xref py py-class docutils literal notranslate"><span class="pre">list</span></code></a>, <a class="reference internal" href="#tuple" title="tuple"><code class="xref py py-class docutils literal notranslate"><span class="pre">tuple</span></code></a>, <a class="reference internal" href="#range" title="range"><code class="xref py py-class docutils literal notranslate"><span class="pre">range</span></code></a><a class="headerlink" href="#sequence-types-list-tuple-range" title="Link permanente para este título">¶</a></h2>
<p>Existem três tipos básicos de sequência: listas, tuplas e objetos range. Tipos de sequência adicionais adaptados para o processamento de <a class="reference internal" href="#binaryseq"><span class="std std-ref">dados binários</span></a> e <a class="reference internal" href="#textseq"><span class="std std-ref">strings de texto</span></a> são descritos em seções dedicadas.</p>

<p>Sequências do mesmo tipo também suportam comparações. Em particular, tuplas e listas são comparadas lexicograficamente pela comparação de elementos correspondentes. Isso significa que para comparar igualmente, cada elemento deve comparar igual e as duas sequências devem ser do mesmo tipo e possuírem o mesmo comprimento. (Para detalhes completos, veja <a class="reference internal" href="../reference/expressions.html#comparisons"><span class="std std-ref">Comparações</span></a> na referência da linguagem.)</p>

</li>
<li><p>Se <em>i</em> ou <em>j</em> for negativo, o índice será relativo ao fim da sequência <em>s</em>: <code class="docutils literal notranslate"><span class="pre">len(s)</span> <span class="pre">+</span> <span class="pre">i</span></code> ou <code class="docutils literal notranslate"><span class="pre">len(s)</span> <span class="pre">+</span> <span class="pre">j</span></code> será substituído. Mas note que <code class="docutils literal notranslate"><span class="pre">-0</span></code> ainda será <code class="docutils literal notranslate"><span class="pre">0</span></code>.</p></li>
<li><p>A fatia <em>s</em> de <em>i</em> até <em>j</em> é definida como a sequência de itens com índice <em>k</em> tal que <code class="docutils literal notranslate"><span class="pre">i</span> <span class="pre">&lt;=</span> <span class="pre">k</span> <span class="pre">&lt;</span> <span class="pre">j</span></code>. Se <em>i</em> ou <em>j</em> foi maior do que <code class="docutils literal notranslate"><span class="pre">len(s)</span></code>, usa <code class="docutils literal notranslate"><span class="pre">len(s)</span></code>. Se <em>i</em> for omitido ou <code class="docutils literal notranslate"><span class="pre">None</span></code>, use <code class="docutils literal notranslate"><span class="pre">0</span></code>. Se <em>j</em> for omitido ou <code class="docutils literal notranslate"><span class="pre">None</span></code>, usa <code class="docutils literal notranslate"><span class="pre">len(s)</span></code>. Se <em>i</em> for maior ou igual a <em>j</em>, a fatia é vazia.</p></li>
<li><p>A fatia <em>s</em> de <em>i</em> até <em>j</em> com passo <em>k</em> é definida como sendo a sequência de itens com índice <code class="docutils literal notranslate"><span class="pre">x</span> <span class="pre">=</span> <span class="pre">i</span> <span class="pre">+</span> <span class="pre">n*k</span></code> tal que <code class="docutils literal notranslate"><span class="pre">0</span> <span class="pre">&lt;=</span> <span class="pre">n</span> <span class="pre">&lt;</span> <span class="pre">(j-i)/k</span></code>. Em outras palavras, os índices são <code class="docutils literal notranslate"><span class="pre">i</span></code>, <code class="docutils literal notranslate"><span class="pre">i+k</span></code>, <code class="docutils literal notranslate"><span class="pre">i+2*k</span></code>, <code class="docutils literal notranslate"><span class="pre">i+3*k</span></code> e assim por diante, parando quando <em>j</em> for atingido (mas nunca incluindo <em>j</em>). Quando <em>k</em> for positivo, <em>i</em> e <em>j</em> serão reduzidos a <code class="docutils literal notranslate"><span class="pre">len(s)</span></code> se forem maiores. Quando <em>k</em> for negativo, <em>i</em> e <em>j</em> são reduzidos para <code class="docutils literal notranslate"><span class="pre">len(s)</span> <span class="pre">-</span> <span class="pre">1</span></code> se forem maiores. Se <em>i</em> ou <em>j</em> forem omitidos ou <code class="docutils literal notranslate"><span class="pre">None</span></code>, eles se tornam valores “finais” (cujo final depende de <em>k</em>). Nota: <em>k</em> não pode ser zero. Se <em>k</em> for <code class="docutils literal notranslate"><span class="pre">None</span></code>, o mesmo será tratado como sendo igual a <code class="docutils literal notranslate"><span class="pre">1</span></code>.</p></li>
<li><p>Concatenar sequências imutáveis sempre resulta em um novo objeto. Isso significa que a criação de uma sequência por concatenação repetida terá um custo quadrático de tempo de execução no comprimento total da sequência. Para obter um custo de tempo de execução linear, devemos alternar para uma das alternativas abaixo:</p>
<ul class="simple">
<li><p>Se estiver concatenando objetos <a class="reference internal" href="#str" title="str"><code class="xref py py-class docutils literal notranslate"><span class="pre">str</span></code></a>, você pode criar uma lista e usar <a class="reference internal" href="#str.join" title="str.join"><code class="xref py py-meth docutils literal notranslate"><span class="pre">str.join()</span></code></a> no final ou então escrever numa instância de <a class="reference internal" href="io.html#io.StringIO" title="io.StringIO"><code class="xref py py-class docutils literal notranslate"><span class="pre">io.StringIO</span></code></a> e recuperar o seu valor ao final</p></li>
<li><p>Se estiver concatenando objetos <a class="reference internal" href="#bytes" title="bytes"><code class="xref py py-class docutils literal notranslate"><span class="pre">bytes</span></code></a>, você também pode usar o método <a class="reference internal" href="#bytes.join" title="bytes.join"><code class="xref py py-meth docutils literal notranslate"><span class="pre">bytes.join()</span></code></a> ou <a class="reference internal" href="io.html#io.BytesIO" title="io.BytesIO"><code class="xref py py-class docutils literal notranslate"><span class="pre">io.BytesIO</span></code></a>, ou você pode fazer concatenação interna com um objeto <a class="reference internal" href="#bytearray" title="bytearray"><code class="xref py py-class docutils literal notranslate"><span class="pre">bytearray</span></code></a>. A classe <a class="reference internal" href="#bytearray" title="bytearray"><code class="xref py py-class docutils literal notranslate"><span class="pre">bytearray</span></code></a> são objetos mutáveis e possuem um eficiente mecanismo de superalocação</p></li>
<li><p>Se estiver concatenando objetos <a class="reference internal" href="#tuple" title="tuple"><code class="xref py py-class docutils literal notranslate"><span class="pre">tuple</span></code></a>, estenda a classe <a class="reference internal" href="#list" title="list"><code class="xref py py-class docutils literal notranslate"><span class="pre">list</span></code></a> em vez disso</p></li>
<li><p>para outros tipos, busque na documentação relevante da classe</p></li>
</ul>
</li>
<li><p>Alguns tipos sequências (como <a class="reference internal" href="#range" title="range"><code class="xref py py-class docutils literal notranslate"><span class="pre">range</span></code></a>) apenas suportam sequências de itens que seguem padrões específicos e, portanto, não suportam concatenação ou repetição de sequência.</p></li>
<li><p><code class="docutils literal notranslate"><span class="pre">index</span></code> levanta <a class="reference internal" href="exceptions.html#ValueError" title="ValueError"><code class="xref py py-exc docutils literal notranslate"><span class="pre">ValueError</span></code></a> quando <em>x</em> não é encontrado em <em>s</em>. Nem todas as implementações suportam a passagem dos argumentos adicionais <em>i</em> e <em>j</em>. Esses argumentos permitem a pesquisa eficiente de subseções da sequência. Passar os argumentos extras é aproximadamente equivalente a usar <code class="docutils literal notranslate"><span class="pre">s[i:j].index(x)</span></code>, apenas sem copiar nenhum dado e com o índice retornado sendo relativo ao início da sequência e não ao início da fatia.</p>

<p>Notas:</p>
<ol class="arabic">
<li><p><em>t</em> deve ter o mesmo comprimento que a fatia a qual ele está substituindo.</p></li>
<li><p>O argumento opcional <em>i</em> tem como  padrão <code class="docutils literal notranslate"><span class="pre">-1</span></code>, de modo que, por padrão, o último item é removido e retornado.</p></li>
<li><p><code class="xref py py-meth docutils literal notranslate"><span class="pre">remove()</span></code> levanta <a class="reference internal" href="exceptions.html#ValueError" title="ValueError"><code class="xref py py-exc docutils literal notranslate"><span class="pre">ValueError</span></code></a> quando <em>x</em> não é encontrado em <em>s</em>.</p></li>
<li><p>O método <code class="xref py py-meth docutils literal notranslate"><span class="pre">reverse()</span></code> modifica a sequência no lugar para economizar espaço ao reverter uma grande sequência. Para lembrar os usuários que isso ocorre como sendo um efeito colateral, o mesmo não retorna a sequência invertida.</p></li>
<li><p><code class="xref py py-meth docutils literal notranslate"><span class="pre">clear()</span></code> e <code class="xref py py-meth docutils literal notranslate"><span class="pre">copy()</span></code> estão incluídos para consistência com as interfaces de contêineres mutáveis que não suportam operações de fatiamento (como <a class="reference internal" href="#dict" title="dict"><code class="xref py py-class docutils literal notranslate"><span class="pre">dict</span></code></a> e <a class="reference internal" href="#set" title="set"><code class="xref py py-class docutils literal notranslate"><span class="pre">set</span></code></a>). <code class="xref py py-meth docutils literal notranslate"><span class="pre">copy()</span></code> não faz parte da ABC de <a class="reference internal" href="collections.abc.html#collections.abc.MutableSequence" title="collections.abc.MutableSequence"><code class="xref py py-class docutils literal notranslate"><span class="pre">collections.abc.MutableSequence</span></code></a>, mas a maioria das classes concretas de sequências mutáveis fornece isso.</p>

<p><span class="versionmodified added">Novo na versão 3.3: </span>Métodos <code class="xref py py-meth docutils literal notranslate"><span class="pre">clear()</span></code> e <code class="xref py py-meth docutils literal notranslate"><span class="pre">copy()</span></code>.</p>
</div>
</li>
<li><p>O valor <em>n</em> é um inteiro, ou um objeto implementando <a class="reference internal" href="../reference/datamodel.html#object.__index__" title="object.__index__"><code class="xref py py-meth docutils literal notranslate"><span class="pre">__index__()</span></code></a>. Valores zero e negativos de <em>n</em> limparão a sequência. Os itens na sequência não são copiados; eles são referenciados várias vezes, como explicado para <code class="docutils literal notranslate"><span class="pre">s</span> <span class="pre">*</span> <span class="pre">n</span></code> em <a class="reference internal" href="#typesseq-common"><span class="std std-ref">Operações comuns de sequências</span></a>.</p>
